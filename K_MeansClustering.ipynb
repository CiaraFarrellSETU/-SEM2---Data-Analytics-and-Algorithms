This notebook looks at K-means clustering.  
K-means clustering is a type of unsupervised learning, this type of machine learning gains an undertsanding of the data without any supervision.  
A key feature of unsuprvised learning is that the data does not need to be split to train the algorithm.

In k-means clustering the following variables can be changed.  
* n_samples: The number of points divided
amoungst the clusters, it can be any integer number.    
*cluster_std: which is the standard deviation of the clusters values closer to 0 make the points more localised around the cluster centre and values closer to 1 make the points more spread out.  
*random_state: this chooses the number generation randomly depending on the value chosen, it can be any integer number.  
*s: the size the blobs are plotted on the graph  
*centres: The number of centers to be generated, it can be any integer number. This is the main element in k-mean clustering as clusters are used to group data points based on their proximity to the cluster centre.

Expectation-Maximization is a specil case of k-means, it is known to converge.  
It works similarly to kmeans clustering except that the data is data is allocated to a cluster not the other way around.  
It guess the cluster centers and then assignes points to the cluster based on proximity.  
By using Jakes code it is possible to plot some random points and randomly chose clusters to go along with it.  
The data points are then assigned to a cluster based on its distance to the cluster centre.  
Instead of starting with 4 clusters, I begin with 2.  
This clustering is not the best as while the top and bottom cluster are nicely done, the middel section of the graph is not well clustered.

```python
from sklearn.metrics import pairwise_distances_argmin

def find_clusters(X, n_clusters, rseed=2):
    # 1. Randomly choose clusters
    rng = np.random.RandomState(rseed)
    i = rng.permutation(X.shape[0])[:n_clusters]
    centers = X[i]

    while True:
        # 2a. Assign labels based on closest center
        labels = pairwise_distances_argmin(X, centers)

        # 2b. Find new centers from means of points
        new_centers = np.array([X[labels == i].mean(0)
                                for i in range(n_clusters)])

        # 2c. Check for convergence
        if np.all(centers == new_centers):
            break
        centers = new_centers

    return centers, labels

centers, labels = find_clusters(X, 2)
plt.scatter(X[:, 0], X[:, 1], c=labels,
            s=20, cmap='jet');
```

This section of the notebook take what i have learned previously and is addapted for a dataset with no predefined cluster.  
By using both the elbow method and silhouette method the exact number of clusters can be determined.  
  
Here 850 random blobs are generated.  
They have the defualt number of featues which is set to two and a random state of 0.  
The blobs are printed at size 10 below.  
One thing that is for sure about this plot is that it there is not clearly distinguishable clusters just by using your eyes.


```python
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=850,n_features=2,
                       cluster_std=1, random_state=0)
plt.scatter(X[:, 0], X[:, 1], s=10);
```

Firstly using the elbow method.  
The elbow method works by using the Within-Cluster Sum of Squares(wcss) which is the sum of the square distances betwwen the cluster points and its centroid.  
By using the elbow method the wcss is set from range of 1 to 9 clusters.  
Code for the elbow method was adapted from https://www.analyticsvidhya.com/blog/2021/01/in-depth-intuition-of-k-means-clustering-algorithm-in-machine-learning/.  


```python
from sklearn.cluster import KMeans
# Within cluster Squared Sum
wcss = []

# Using Elbow method to determine
# the proper n(no of clusters)
for i in range(1, 10):
    kmeans = KMeans(n_clusters=i,
                    init='k-means++',
                    random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

Now the eblow mthod will be plotted from the wcss range of 1 to 10.  
However there is not a very clear elbow point on the graph it could potentially be 3 as it has the sharped drop in values but it could realistically be any cluster values from 2 to 6.

```python
plt.plot(range(1, 10), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

I tried printing the wcss to try give an easier way to see the drop

```python
print(wcss)
```

Now I attempt to find the best number of clusters using the silhouette method.  
The code for the creation of the silhouette plot was taken from https://scikit-learn.org/stable/auto_examples/cluster/plot_kmeans_silhouette_analysis.html .  
As from the elbow method a range of 2 to 6 could be optinal cluster sizes so that is the set range_n_clusters in the code.  
The code the created the corresponding silhouette plots for each cluster size.  
The average silhouette score for each cluster is also printed.  
A score closer to 1 denotes the best cluster size.



```python
from sklearn.metrics import silhouette_samples, silhouette_score
import matplotlib.cm as cm
range_n_clusters = [2, 3, 4, 5, 6]

for n_clusters in range_n_clusters:
    # Create a subplot with 1 row and 2 columns
    fig, (ax1, ax2) = plt.subplots(1, 2)
    fig.set_size_inches(18, 7)

    # The 1st subplot is the silhouette plot
    # The silhouette coefficient can range from -1, 1 but in this example all
    # lie within [-0.1, 1]
    ax1.set_xlim([-0.1, 1])
    ax1.set_ylim([0, len(X) + (n_clusters + 1) * 10])

    clusterer = KMeans(n_clusters=n_clusters, random_state=10)
    cluster_labels = clusterer.fit_predict(X)

    # The silhouette_score gives the average value for all the samples.
    # This gives a perspective into the density and separation of the formed
    # clusters
    silhouette_avg = silhouette_score(X, cluster_labels)
    print(
        "For n_clusters =",
        n_clusters,
        "The average silhouette_score is :",
        silhouette_avg,
    )

    # Compute the silhouette scores for each sample
    sample_silhouette_values = silhouette_samples(X, cluster_labels)

    y_lower = 10
    for i in range(n_clusters):
        # Aggregate the silhouette scores for samples belonging to
        # cluster i, and sort them
        ith_cluster_silhouette_values = sample_silhouette_values[cluster_labels == i]

        ith_cluster_silhouette_values.sort()

        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i

        color = cm.nipy_spectral(float(i) / n_clusters)
        ax1.fill_betweenx(
            np.arange(y_lower, y_upper),
            0,
            ith_cluster_silhouette_values,
            facecolor=color,
            edgecolor=color,
            alpha=0.7,
        )

        # Label the silhouette plots with their cluster numbers at the middle
        ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

        # Compute the new y_lower for next plot
        y_lower = y_upper + 10  # 10 for the 0 samples

    ax1.set_title("The silhouette plot for the various clusters.")
    ax1.set_xlabel("The silhouette coefficient values")
    ax1.set_ylabel("Cluster label")

    # The vertical line for average silhouette score of all the values
    ax1.axvline(x=silhouette_avg, color="red", linestyle="--")

    ax1.set_yticks([])  # Clear the yaxis labels / ticks
    ax1.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])

    # 2nd Plot showing the actual clusters formed
    colors = cm.nipy_spectral(cluster_labels.astype(float) / n_clusters)
    ax2.scatter(
        X[:, 0], X[:, 1], marker=".", s=30, lw=0, alpha=0.7, c=colors, edgecolor="k"
    )

    # Labeling the clusters
    centers = clusterer.cluster_centers_
    # Draw white circles at cluster centers
    ax2.scatter(
        centers[:, 0],
        centers[:, 1],
        marker="o",
        c="white",
        alpha=1,
        s=200,
        edgecolor="k",
    )

    for i, c in enumerate(centers):
        ax2.scatter(c[0], c[1], marker="$%d$" % i, alpha=1, s=50, edgecolor="k")

    ax2.set_title("The visualization of the clustered data.")
    ax2.set_xlabel("Feature space for the 1st feature")
    ax2.set_ylabel("Feature space for the 2nd feature")

    plt.suptitle(
        "Silhouette analysis for KMeans clustering on sample data with n_clusters = %d"
        % n_clusters,
        fontsize=14,
        fontweight="bold",
    )
```
